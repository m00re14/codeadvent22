```python
with open ('input day 2.txt', 'r') as input:
    input_list = input.read().splitlines()
    
total_score1 = 0
total_score2 = 0

#rock = 1
#paper = 2
#scissors = 3

def part_one(opponent, you):
    round_score = 0
    if opponent == 'A': #Rock
        
        if you == 'Y':
            round_score += 8 #Win(Paper)
        
        if you == 'X':
            round_score += 4 #Draw(Rock)
        
        if you == 'Z':
            round_score += 3 #Loss(Scissors)
            
            
    elif opponent == 'B': #Paper
        
        if you == 'Z':
            round_score += 9 #Win(Scissors)
            
        if you == 'Y':
            round_score += 5 #Draw(Paper)
            
        if you == 'X':
            round_score += 1 #Loss(Rock)
            
    elif opponent == 'C': #Scissors
        
        if you == 'X':
            round_score += 7 #Win(Rock)
        
        if you == 'Z':
            round_score += 6 #Draw(Scissors)
        
        if you == 'Y':
            round_score += 2 #Draw(Paper)
    return round_score


def part_two(opponent, you):
    round_score = 0
    if opponent == 'A': #Rock
        
        if you == 'X':
            round_score += 3 #Need Loss
            
        if you == 'Y':
            round_score += 4 #Need Draw
        
        if you == 'Z':
            round_score += 8 #Need Win
            
            
    elif opponent == 'B': #Paper
        
        if you == 'X':
            round_score += 1 #Need Loss
            
        if you == 'Y':
            round_score += 5 #Need Draw
        
        if you == 'Z':
            round_score += 9 #Need Win
            
    elif opponent == 'C': #Scissors
        
        if you == 'X':
            round_score += 2 #Need Loss
            
        if you == 'Y':
            round_score += 6 #Need Draw
        
        if you == 'Z':
            round_score += 7 #Need Win
    return round_score



for input in input_list:
    total_score1 += part_one(input[0], input[2])
    total_score2 += part_two(input[0], input[2])

print(total_score1)
print(total_score2)
            
        
        
        
            
    
```
